```python
# Import Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import warnings
warnings.filterwarnings('ignore')
```

```python
# Loading the dataset to pandas dataframe
credit_card_data = pd.read_csv('creditcard.csv')
```

```python
# first five rows
credit_card_data.head()
```

```python
# last five rows
credit_card_data.tail()
```

```python
# dataset information
credit_card_data.info()
```

```python
# checking the number of missing values
credit_card_data.isnull().sum().sum()
```

```python
# distribution of legit transactions & fraudulent transactions
credit_card_data['Class'].value_counts()
```

This dataset is highly unbalanced.

0 --> Normal Transactions
1 --> Fraudulent Transactions

```python
# septarating data for analysis
legit = credit_card_data[credit_card_data['Class']==0]
fraud = credit_card_data[credit_card_data['Class']==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# Statistical Measures of the data
legit['Amount'].describe()
```

```python
fraud['Amount'].describe()
```

```python
# Compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

Under-Sampling

Build a sample dataset containing similar distribution of normal transactions and fraudulent transactions

Number of fraudulent transactions --> 492

```python
legit_sample = legit.sample(n=492,random_state=0)
```

```python
# Concatenating two dataframes
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

The difference is still there. Therefore basic nature of data has not changed.

```python
# Splitting the data into Features and Targets
X = new_dataset.iloc[:,0:-1]
Y = new_dataset.iloc[:, -1]
```

```python
X.head()
```

```python
Y.head()
```

```python
# Splitting the data into training data and testing data
x_train, x_test , y_train, y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
# Model Creation

model = LogisticRegression()
```

```python
# Training the model
model.fit(x_train, y_train)
```

```python
# Predicting values based on testing data

y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
# Model Evaluation

accuracy_score(y_test, y_pred)
```

Accuracy Score is 93.40%.

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

Model Precision --> 96%

Model Recall --> 91%

Accuracy, Precision and Recall, all are above 90% which demonstrates that this model is highly accurate in identifying the legit and fraudulent credit card transactions.
